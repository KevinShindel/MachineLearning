### Ідея:

> Модель намагається розкласти взаємодії між фічами на латентні (приховані) фактори.

👉 Ключова проблема, яку вони вирішують:

- багато категоріальних змінних
- sparse дані (наприклад user–item)
- складні взаємодії (feature × feature)

## Де використовують ?

- recommender systems
- ads ranking
- CTR prediction

## Інтуїція

Замість того щоб вчити взаємодію напряму:

> user_id × item_id → rating

модель каже:

давай представимо user і item як вектори (embeddings)

> user → vector (k)
> item → vector (k)
>
> prediction = dot(user_vector, item_vector)


## Типи моделей

1. Matrix Factorization
2. Factorization Machines (FM)
3. Field-aware FM (FFM)

```python
from utils import Smote
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, balanced_accuracy_score, accuracy_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
```

```python
# Create an imbalanced dataset
X, y = make_classification(n_samples=5000, n_features=2, n_informative=2,
                           n_redundant=0, n_repeated=0, n_classes=3,
                           n_clusters_per_class=1,
                           weights=[0.01, 0.05, 0.94],
                           class_sep=0.8, random_state=0)
```

```python
# Train a logistic regression model on the original imbalanced dataset
model = LogisticRegression(solver='lbfgs', max_iter=1000, random_state=0)

model.fit(X, y)
y_pred = model.predict(X)

print("Classification Report (Original):")
print(classification_report(y, y_pred))
```

```python
# calculate balanced accuracy and AUC-ROC for original dataset
balanced_acc = balanced_accuracy_score(y, y_pred)
roc_auc = roc_auc_score(y, model.predict_proba(X), multi_class='ovr')
print(f"Balanced Accuracy (Original): {balanced_acc:.4f}")
print(f"AUC-ROC (Original): {roc_auc:.4f}")
```

```python
# Apply SMOTE to balance the dataset
smote = Smote(k_neighbors=5)
X_resampled, y_resampled = smote.fit_resample(X, y)
# Train a logistic regression model on the resampled dataset
model_resampled = LogisticRegression(solver='lbfgs', max_iter=1000, random_state=0)
model_resampled.fit(X_resampled, y_resampled)
y_pred_resampled = model_resampled.predict(X_resampled)

print("Classification Report (SMOTE Resampled):")
print(classification_report(y_resampled, y_pred_resampled))
```

```python
# calculate balanced accuracy and AUC-ROC for resampled dataset
balanced_acc_resampled = balanced_accuracy_score(y_resampled, y_pred_resampled)
roc_auc_resampled = roc_auc_score(y_resampled, model_resampled.predict_proba(X_resampled), multi_class='ovr')
print(f"Balanced Accuracy (SMOTE Resampled): {balanced_acc_resampled:.4f}")
print(f"AUC-ROC (SMOTE Resampled): {roc_auc_resampled:.4f}")
```
